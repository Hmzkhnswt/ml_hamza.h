```python
import pandas as pd
import numpy as np
import seaborn as sns
```

### Iris Dataset from Sklearn

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.feature_names)
print(iris.target)
```

```python
x = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
trainx,testx,trainy,testy = train_test_split(x,y,test_size=0.2, random_state=41)
```

```python
print(trainx.shape)
print(trainy.shape)
print(testx.shape)
print(testy.shape)
```

### Hyperparameters Tuning 

```python
from sklearn.tree import DecisionTreeClassifier
clf_model = DecisionTreeClassifier(criterion='gini', splitter='random', max_depth=4, min_samples_split=25)
```

```python
clf_model.fit(trainx, trainy)
```

```python
y_pred = clf_model.predict(testx)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf_model)
```

### Tree Visualization

```python
import graphviz.backend as be
from dtreeviz import *
from IPython.display import Image, display_svg, SVG

viz = dtreeviz(clf_model, 
                     trainx,
                     trainy,
                     feature_names=iris.feature_names, 
                     class_names=["setosa", "versicolor", "virginica"])

viz
```

### Checking Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(testy,y_pred)
```

### Regression Trees while Importing data from Sklearn

```python
from sklearn.datasets import make_regression
X,y = make_regression(n_samples=250, n_features=4, n_targets=1, noise=0.2, random_state=41)
```

```python
print(X.shape)
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=41)
```

### Modeling

```python
from sklearn.tree import DecisionTreeRegressor
model_tree_regr = DecisionTreeRegressor( criterion='squared_error')
```

```python
model_tree_regr.fit(x_train, y_train)
```

### Getting tree Structure

```python
from sklearn.tree import plot_tree
plot_tree(model_tree_regr)
```

```python
y_hat = model_tree_regr.predict(x_test)
```

### Accuracy Metrics

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_hat)
```
